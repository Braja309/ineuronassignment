```python
#1->-
def find_the_difference(String1,String2):
    '''It will take out letter from String2, which is not present in String1'''
    for i in String2:
        if String1.count(i) != String2.count(i):
            return i
String1 = input("Please enter the first string : ")
String2 = input("Please enter the second string : ")
print("The difference letter is :",find_the_difference(String1,String2))
```

```python
#2->
def abc(args):
    '''It will return count of each data types from given arguments'''
    List = [0,0,0,0,0,0]
    for i in args:
        if type(i) == int:
            List[0] +=1
        elif type(i) == str:
            List[1] +=1
        elif type(i) == bool:
            List[2] +=1
        elif type(i) == list:
            List[3] +=1
        elif type(i) == tuple:
            List[4] +=1
        elif type(i) == dict:
            List[5] +=1
    return List
args = []
while True:
    Inp = input("Do you want to put more arguments 'Yes/No' : ")
    if Inp == 'Yes':
        try:
            i = input("Please enter the argument : ")
            args.append(eval(i))
        except NameError:
            args.append(i)
    else:
        break
print("List of count of datatype is :",abc(args))
```

```python
#3->
def fib_str(count, List):
    '''It will return count of fibonacci string starting with 2string given in list'''
    for i in range(2,count):
        List.append(List[i-1]+List[i-2])
    return ', '.join(List)
Count = int(input("Please enter the number of fibonacci series : "))
List = eval(input("Please enter first two string in a list : "))
print("Fibonacci sequence is :",fib_str(Count,List))
```

```python
#4->
class number_count:
    def __init__(self,num):
        self.num = num
    def ones_threes_nines(self):
        '''It will return a string with number of nines, threes, and ones'''
        num2words = {1: 'Ones', 3: 'Threes', 9: 'Nines'}
        List = []
        for i in [9,3,1]:
            List.append(num2words[i]+':'+str(self.num//i))
            self.num %= i
        self.answer = ', '.join(List)
        return self.answer
Num = int(input("Please enter the number : "))
Num_count = number_count(Num)
print("Answer is :",Num_count.ones_threes_nines())
```

```python
#5->
def fib(seq):
    """It will return fibonacci number on seq"""
    a = 0
    b = 1
    for i in range(seq):
        a,b = b,a+b
    return a
num = int(input("Please enter the sequence : "))
print("Fibonacci number is :",fib(num))
```
