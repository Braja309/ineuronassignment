```python
#3->
l=[]
def dictval(a):
    """It will return list of all the value from a dictionary, even if it is nested dictionary"""
    for i in a.values():
        if type(i)==dict:
            dictval(i)
        else:
            l.append(i)
    return l
print(dictval({'a':2,'b':{'c':3,'d':4},'e':5}))
```

```python
#5->
def con_list(*kargs):
    """It will return concatenated list of given multiple lists"""
    list = []
    for i in kargs:
        for j in i:
            list.append(j)
    return list
print(con_list([1,2,3],[23,4,5]))
```

```python
#11->
import cv2
def readVideo(file):
    """It will read an video file and play it"""
    video = cv2.VideoCapture(file)
    if video.isOpened()==False:
        print("error opening the file")
    while video.isOpened():
        flag,frame = video.read()
        if flag==True:
            cv2.imshow('Frame',frame)
            if cv2.waitKey(25) & 0xFF == ord('b'):
                break
        else:
            break
    video.release()
    cv2.destroyAllWindows()
readVideo("D:\\movies\\Bengali\\Sonar Kella\\Sonar Kella 1974.avi")
```

```python
#16->
import PyPDF2
def pdfread(file):
    """It will read complete pdf file"""
    pdfFileObj = open(file, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    #print(pdfReader.numPages)
    text = ''
    for i in range(pdfReader.numPages):
        pageObj = pdfReader.getPage(i)
        text = text+pageObj.extractText()
    pdfFileObj.close()
    return text
file = "C:\\Users\\Braja\\Desktop\\VOD_Virtual Desktops Access Guide.pdf"
print(pdfread(file))
```

```python
#17->
import docx
def docread(file):
    """It will read document file"""
    doc = docx.Document(file)
    text = ''
    for i in doc.paragraphs:
        text = text+i.text+'\n'
    return text
file = "C:\\Users\\Braja\\Desktop\\Data Science\\ineuron Class\\Assignments\\Python\\Basic\\Assignment_1_Basic.docx"
print(docread(file))  
```

```python
#18->
import os
folder = "C:\\Users\\Braja\\Desktop\\Data Science\\ineuron Class\\Assignments\\Python\\Basic\\AssignmentTest"
extensions = ['.doc','.docx']
def docfiles(folder,extensions):
    """It will filter out only specific extension files"""
    List = []
    for dirpath, subfolder, filenames in os.walk(folder):
        for filename in filenames:
            name, ext = os.path.splitext(filename)
            if ext in extensions:
                List.append(filename)
        break
    return List
print(docfiles(folder,extensions))
```
